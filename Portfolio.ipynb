### 2. Inverted Pendulum on Cart

The classic control theory test is balancing an inverted pendulum. I took on the challenge to learn about several things including state-space control and mechanical design. Through many hours of tuning I gained a deep appreciation for the mathematics of control. Despite its short-comings, I wish to present this project to you because of the sheer willpower it took to bring it into existence. What you see below is roughly 9 weeks of late nights, working my hardest to implement state-space control, nonlinear Lyapunov control, mechanical design, ODrive motor driver, software (C/C++, Python, Bash), and electrical design.

Thorough documentation is available below. Please let me know if you have any questions.

  * Documentation - https://github.com/vs65497/InvertedPendulum/blob/main/README.md

### 3. Research: Slothbot

The Slothbot is a research project under Magnus Egerstedt, Dean of Engineering at UC Irvine. This project in particular is the 3rd edition in a line of Slothbots beginning in the GRITS lab at Georgia Tech. The goal is for it to perform long-duration environmental monitoring while traversing a high-wire. Below, you will find documentation of a complex software-electrical system which allows the Slothbot to hibernate and conserve power for long periods of time. It involves an onboard computer, on board microcontroller, and base-station computer which communicates with a server. It builds on my understanding of full-stack websites, IoT devices, and mobile robots.

Formal documentation for this project is still in progress. More information is available on request.

  *  Documentation - https://drive.google.com/file/d/1GCOMs7PnJ58wxgqbabNIW4C42T3pKNw9/view?usp=drive_link

### 4. Tactile Sensor Experiment

Robotics has traditionally been a control field. Many of the algorithms are based on optimization which attempt to tightly control trajectories. More recently machine learning has made its way onto the scene. This project serves 2 purposes:

(A) To study machine learning. Particularly YOLOv5 (computer vision) and tabular prediction via a single layer deep neural network.

(B) To attempt creating a rudimentary tactile sensor which would allow a robotic manipulator to sense an object it is grasping.

This project demonstrates a complex stack of technology. Not only is the final system multilayered, but it also features a complex construction process. More information is available in the documentation link. Please feel free to enquire further.

  * Documentation - https://github.com/vs65497/Tactile-Sensor
